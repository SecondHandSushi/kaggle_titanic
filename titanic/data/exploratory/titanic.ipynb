```python
import hvplot.pandas
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
titanic_raw_df=pd.read_csv('../titanic.csv')
titanic_raw_df.head()
```

```python
titanic_raw_df.info()
```

```python
titanic_column_datatype = {
    'Name': 'string',
    'Sex': 'string',
    'Ticket': 'string',
    'Cabin' : 'string',
    'Embarked' : 'string'
}
titanic_df = titanic_raw_df.astype(titanic_column_datatype)
titanic_df['Fare'] = titanic_df['Fare'].round(2)
```

```python
titanic_df = titanic_df.copy()

# Drop rows where all values in 'Cabin', 'Age', and 'Embarked' columns are missing
titanic_df.dropna(subset=['Cabin', 'Age', 'Embarked'], how='all', inplace=True)

# Display the first few rows of the modified DataFrame
titanic_df.head()
```

```python
titanic_no_cab_tick = titanic_df.copy()
titanic_no_cab_tick.drop(columns=['Cabin','Ticket'], inplace=True)
titanic_no_cab_tick=titanic_no_cab_tick.set_index(['Name'])
titanic_no_cab_tick.head()
```

```python
titanic_no_cab_tick_cat_col=['Sex','Embarked']
titanic_no_cab_tick = pd.get_dummies(titanic_no_cab_tick, columns=titanic_no_cab_tick_cat_col)
titanic_no_cab_tick.head()

```
