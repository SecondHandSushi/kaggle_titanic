```python
import hvplot.pandas
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
train_raw_df=pd.read_csv('../train.csv')
train_raw_df.head()
```

```python
train_raw_df.info()
```

```python
train_column_datatype = {
    'Name': 'string',
    'Sex': 'string',
    'Ticket': 'string',
    'Cabin' : 'string',
    'Embarked' : 'string'
}
train_df = train_raw_df.astype(train_column_datatype)
train_df['Fare'] = train_df['Fare'].round(2)
```

```python
train_df = train_df.copy()
train_df.dropna(subset=['Cabin', 'Age', 'Embarked'], how='all', inplace=True)
train_df.head()
```

```python

train_df = train_df[train_df['Age'].notna() | train_df['Embarked'].notna()]
train_df.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(train_df):
    # Filter rows based on column: 'Embarked'
    train_df = train_df[train_df['Cabin'].notna()]
    # Filter rows based on column: 'Embarked'
    train_df = train_df[train_df['Embarked'].notna()]
    # Filter rows based on column: 'Age'
    train_df = train_df[train_df['Age'].notna()]
    return train_df

train_df_clean = clean_data(train_df.copy())
train_df_clean.head()
```

```python
train_df_clean = train_df_clean.sort_values(by='Fare', ascending=False)
train_df_clean.head()
```
